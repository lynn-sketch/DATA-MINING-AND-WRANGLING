```python
##Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
##Reading the dataset 
testData = pd.read_csv('car_price_dataset.csv')
testData.head(10)
```

```python
testData.isna().sum()
```

```python
##Creating categorical data
num_data = testData.select_dtypes(exclude='object')
num_data
```

```python
##Checking for outliers
def check_outliers(data):
    plt.figure(figsize=(20,20))
    for i,col in enumerate(data.columns):
        plt.subplot(3,3,i+1)
        sns.boxplot(data[col])
        plt.title(col)
plt.tight_layout()
plt.show()  

```

```python
##Plotting the numerical data
check_outliers(num_data)
```
